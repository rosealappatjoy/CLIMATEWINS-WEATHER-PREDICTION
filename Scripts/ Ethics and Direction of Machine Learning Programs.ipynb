## Scaling Data

### Create a scaler object using StandardScaler from sklearn.preprocessing. This scaler assumes that the data follows a normal distribution and scales each feature so that it has a mean of 0 and a standard deviation of 1. Each variable is scaled independently from the others

```python
scaler = StandardScaler()
```

```python
#create new df with scaled data

df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled.head()
```

```python
# Export dataframe as .csv
df_scaled.to_csv(os.path.join(path, 'DataSet_scaled.csv'))
```
